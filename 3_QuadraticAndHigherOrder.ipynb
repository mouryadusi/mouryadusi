```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define Quadratic Parameters ---
# We'll create data for the equation: y = b2*x^2 + b1*x + b0
b2 = 2.5  # formerly 'a'
b1 = -3.0 # formerly 'b'
b0 = 10.0 # formerly 'c'

# --- 2. Generate Data ---
# Create 100 x-values from -10 to 10
x_data = np.linspace(-10, 10, 100)

# Calculate the corresponding y-values based on the quadratic equation
y_true = b2 * x_data**2 + b1 * x_data + b0
```

```python

# Add some random "noise" to the y-values to simulate real-world data
# This makes the fitting process more interesting.
noise = np.random.normal(0, 15, x_data.shape) # Mean=0, StdDev=15
y_data = y_true + noise

# --- 3. Plot the Data ---
# Create a scatter plot of the generated data
plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, label='Generated Noisy Data', color='blue', alpha=0.7)

# --- 4. Customize and Show Plot ---
plt.title('Scatter Plot of Generated Quadratic Data', fontsize=16)
plt.xlabel('X Values', fontsize=12)
plt.ylabel('Y Values', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

```

```python

# --- 5. Plot Noisy Data and True Equation ---
# Create a new figure to show both the noisy data and the true underlying equation
plt.figure(figsize=(10, 6))

# Scatter plot for the noisy data in blue
plt.scatter(x_data, y_data, label='Generated Noisy Data', color='blue', alpha=0.6)

# Line plot for the true equation in red
plt.plot(x_data, y_true, label='True Equation (y_true)', color='red', linewidth=3)

# --- 6. Customize and Show the New Plot ---
plt.title('Generated Data vs. True Equation', fontsize=16)
plt.xlabel('X Values', fontsize=12)
plt.ylabel('Y Values', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

```

```python
# --- 5. Demo for a 6th-Order Polynomial ---
# Manually define coefficients for a 6th order polynomial
# y = b6_6*x^6 + b6_5*x^5 + ... + b6_1*x + b6_0
b6_6 = 0.001   # coefficient for x^6
b6_5 = -0.005  # coefficient for x^5
b6_4 = -0.05   # coefficient for x^4
b6_3 = 0.1     # coefficient for x^3
b6_2 = 1.5     # coefficient for x^2
b6_1 = -2.0    # coefficient for x^1
b6_0 = 8.0     # coefficient for x^0 (intercept)

# Calculate the y-values for the 6th order polynomial
y_poly6 = (b6_6 * x_data**6 +
           b6_5 * x_data**5 +
           b6_4 * x_data**4 +
           b6_3 * x_data**3 +
           b6_2 * x_data**2 +
           b6_1 * x_data +
           b6_0)

# --- 6. Plot the Data and the High-Order Polynomial ---
plt.figure(figsize=(12, 7))

# Plot the original noisy data
plt.scatter(x_data, y_data, label='Original Noisy Data', color='blue', alpha=0.6)

# Plot the 6th-order polynomial curve
plt.plot(x_data, y_poly6, label='6th-Order Polynomial', color='purple', linewidth=2)

# --- 7. Customize and Show the New Plot ---
plt.title('Data vs. Manually Defined 6th-Order Polynomial', fontsize=16)
plt.xlabel('X Values', fontsize=12)
plt.ylabel('Y Values', fontsize=12)
plt.legend()
plt.grid(True)
plt.ylim(min(y_data) - 20, max(y_data) + 20) # Adjust y-axis
plt.show()

```
